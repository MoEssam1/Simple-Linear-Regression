# import the librares

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

# import dataset

```python
dataset = pd.read_csv("Salary_Data.csv")
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

# split dataset into train set and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

# train the simple linear regression model on the traing set 

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

# predicting test set 

```python
y_pred = regressor.predict(x_test)
```

# visualising the training set

```python
plt.scatter(x_train,y_train,color = "red")
plt.plot(x_train,regressor.predict(x_train))
plt.title("visualising the training set")
plt.xlabel("years of exeperance")
plt.ylabel("salary")
plt.show()
```

# visualising the testing set

```python
plt.scatter(x_test,y_test,color = "red")
plt.plot(x_test,y_pred)
plt.title("visualising the testing set")
plt.xlabel("years of exeperance")
plt.ylabel("salary")
plt.show()
```

# predicting one single value

```python
print(regressor.predict([[12]]))
```

# get the final simple linear regression equation

```python
print(regressor.coef_)
print(regressor.intercept_)
```
